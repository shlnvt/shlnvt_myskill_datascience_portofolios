#Library

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#Load data

```python
dataset = sns.load_dataset("mpg")
dataset.head()
```

#Membuat scatterplot

```python
sns.scatterplot(data = dataset,
                x = "weight",
                y = "mpg")
plt.show()
```

```python
sns.scatterplot(data = dataset,
                x = "acceleration",
                y = "mpg")
plt.show()
```

```python
sns.scatterplot(data = dataset,
                x = "displacement",
                y = "mpg")
plt.show()
```

```python
sns.scatterplot(data = dataset,
                x = "displacement",
                y = "horsepower")
plt.show()
```

#Membuat korelasi

```python
korelasi = dataset.corr()
korelasi
```

```python
sns.heatmap(korelasi, vmax = 1, vmin = -1, center = 0)
plt.show()
```

#Uji hipotesis

```python
from scipy.stats import pearsonr
```

```python
coef, p_val = pearsonr(dataset["mpg"], dataset["displacement"])
```
